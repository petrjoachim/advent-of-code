```python
# Star 1
from typing import NamedTuple

route = NamedTuple("Route", [("target", int), ("source", int), ("range", int)])


def parse_map(lines: str) -> list[route]:
    return list(map(lambda _: route(*map(int, _.split(" "))), lines))


def get_destination(source: int, map: list[route]) -> int:
    for route in map:
        if source >= route.source and source <= route.source + route.range:
            target = route.target + (source - route.source)
            # print(source, route, "=>", target)
            return target
    # print('default target', source)
    return source

def star1(input: str) -> int:
    sections = list(map(lambda _: _.strip(), input.split("\n\n")))
    seeds = list(map(int, sections[0].split(": ")[1].split(" ")))

    seed_to_soil = parse_map(sections[1].split("\n")[1:])
    soil_to_fertiliser = parse_map(sections[2].split("\n")[1:])
    fertiliser_to_water = parse_map(sections[3].split("\n")[1:])
    water_to_light = parse_map(sections[4].split("\n")[1:])
    light_to_temperature = parse_map(sections[5].split("\n")[1:])
    temperature_to_humidity = parse_map(sections[6].split("\n")[1:])
    humidity_to_location = parse_map(sections[7].split("\n")[1:])

    soils = map(lambda s: get_destination(s, seed_to_soil), seeds)
    fertilisers = map(lambda s: get_destination(s, soil_to_fertiliser), soils)
    waters = map(lambda s: get_destination(s, fertiliser_to_water), fertilisers)
    lights = map(lambda s: get_destination(s, water_to_light), waters)
    temperatures = map(lambda s: get_destination(s, light_to_temperature), lights)
    humidities = map(lambda s: get_destination(s, temperature_to_humidity), temperatures)
    locations = map(lambda s: get_destination(s, humidity_to_location), humidities)

    return min(locations)


assert (star1(test_input) == 35)
star1(my_input)  # => 346433842
```
