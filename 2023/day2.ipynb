```python
# Star 1

cubes_count = {
    'red': 12,
    'green': 13,
    'blue': 14,
}


def get_valid_games_ids(lines: list) -> list:
    for line in lines:
        line = line.strip()
        game_id, games = line.split(': ')
        game_id = game_id.strip("Game")
        games = games.split(';')
        valid = True
        for game in games:
            cubes = game.strip().split(', ')
            for cube in cubes:
                count, color = cube.split(' ')
                if int(count) > cubes_count[color]:
                    valid = False

        if valid:
            # print(line, "=>", True)
            yield int(game_id)
        # print(line, "=>", False)


def star1(input: str) -> int:
    return sum(get_valid_games_ids(input.split('\n')))


assert (star1(test_input) == 8)
star1(my_input)
```

```python
# Star 2
from collections import defaultdict
from math import prod
Ã¥

def get_game_power(game: str) -> int:
    peeks = game.split(': ')[1].split(';')
    cubes = defaultdict(int)
    for peek in peeks:
        for cube in peek.strip().split(', '):
            count, color = cube.split(' ')
            cubes[color] = max(int(count), cubes[color])

    return prod(cubes.values())


def star2(input: str) -> int:
    return sum(map(get_game_power, input.split('\n')))


star2(test_input)  # => 2286
assert (star2(test_input) == 2286)

star2(my_input)
```
